```python
# population with French as the first language spoken at home. the column type is binary (0 as No, 1 as Yes)
df['FOL']
```

```python
# estimate the population proportion that speaks French as their first official language at home. 95% C.I of the attribute
# french first lang spoken proportion from the sample
fr_p_hat = len(df[df['FOL'] == 2]) / len(df)
fr_p_hat
```

```python
# constructing a 95% confidence interval
lower_bound_fr_pp = fr_p_hat - 1.96 * math.sqrt(fr_p_hat * (1 - fr_p_hat)) / math.sqrt(len(df))
upper_bound_fr_pp = fr_p_hat + 1.96 * math.sqrt(fr_p_hat * (1 - fr_p_hat)) / math.sqrt(len(df))
print(f"95% confidence interval for FOL: ({lower_bound_fr_pp}, {upper_bound_fr_pp})")
```

There are values that are not applicable (99999999 and 88888888). We proceed setting those values to 0.

### Estimate the population mean employment income

```python
# sample mean for employment income
empin_mean_hat = df['EmpIn'].mean()
empin_mean_hat
```

```python
empin_lst = list(df['EmpIn'])
empin_std_err = statistics.stdev(empin_lst) / (len(df) ** 0.5)
empin_std_err     # standard error for employment income
```

```python
# margin of error
margin_of_error_empin = 1.96 * empin_std_err
print(f"Margin of error for EmpIn: {margin_of_error_empin}")         # 1.96 is the Z-value for a 95% confidence interval
```

```python
# 95% confidence interval for the population mean
lower_bound_empin = empin_mean_hat - margin_of_error_empin
upper_bound_empin = empin_mean_hat + margin_of_error_empin

print(f"95% confidence interval for EmpIn: ({lower_bound_empin}, {upper_bound_empin})")
```

### Estimate the population proportion number of Canadian residents between ages of 20 and 29 (binary variable)

```python
# sample proportion of Canadian residents between ages 20 and 29
agegrp_req_p_hat = len(df[(df['AGEGRP'] == 8) | (df['AGEGRP'] == 9)]) / len(df)
agegrp_req_p_hat
```

```python
agegrp_req_lst = list(df[(df['AGEGRP'] == 8) | (df['AGEGRP'] == 9)]['AGEGRP'])
agegrp_n = len(agegrp_req_lst)
n = len(df)
# std_err_agegrp_req = math.sqrt(agegrp_req_p_hat * (1 - agegrp_req_p_hat))
# print(f"Standard error for AGEGRP of Code 8 and 9: {std_err_agegrp_req}")

std_err_agegrp_req = math.sqrt(1 - agegrp_n / n) * math.sqrt(agegrp_req_p_hat * (1 - agegrp_req_p_hat)) / math.sqrt(agegrp_n)
print(f"Standard error for AGEGRP of Code 8 and 9: {std_err_agegrp_req}")
```

```python
# constructing a 95% confidence interval
lower_bound_agegrp_req = agegrp_req_p_hat - 1.96 * math.sqrt(agegrp_req_p_hat * (1 - agegrp_req_p_hat)) / math.sqrt(len(df))
upper_bound_agegrp_req = agegrp_req_p_hat + 1.96 * math.sqrt(agegrp_req_p_hat * (1 - agegrp_req_p_hat)) / math.sqrt(len(df))
print(f"95% confidence interval for AGEGRP of Code 8 and 9: ({lower_bound_agegrp_req}, {upper_bound_agegrp_req})")
```
